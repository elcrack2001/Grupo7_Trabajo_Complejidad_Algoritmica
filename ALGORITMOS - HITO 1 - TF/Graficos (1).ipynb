#Graficos
### Graficos para diferenciar visualmente los tiempos de los algoritmos

```
import random
import time
from copy import copy 
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
```

```
def Grafico(Pdv, PBFS, AyB):
  
  plot(Pdv, 'o-')
  plot(PBFS, 'o-')
  
  plot(Pdv, label='Divide y Venceras')
  plot(PBFS, label='BFS')
  plot(AyB, label='Alfa y Beta')
  xlabel('Cantidad de Pruebas')
  ylabel('Tiempo')
  title('Comparacion de tiempos')
  text(1, 700, 'tabla 9x9 2 peones') 
  text(2, 1000, 'tabla 9x9 4 peones') 
  text(3, 1400, 'tabla 15x15 2 peones') 
  text(4, 2000, ' tabla 15x15 4 peones') 
  text(5, 3000, 'tabla 25x25 2 peones') 
  grid()
  legend(loc='upper left')
  
  show()
```
