#Graficos
### Graficos para diferenciar visualmente los tiempos de los algoritmos

```
import random
import time
from copy import copy 
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
```

```
def Grafico(Pdv, PBFS):
  
  plot(Pdv, 'o-')
  plot(PBFS, 'o-')
  
  plot(Pdv, label='Divide y Venceras')
  plot(PBFS, label='BFS')
  xlabel('Cantidad de Pruebas')
  ylabel('Tiempo')
  title('Comparacion de tiempos')
  text(1, 700, 'tabla 9x9 2 peones') 
  text(2, 1000, 'tabla 9x9 4 peones') 
  text(3, 1400, 'tabla 15x15 2 peones') 
  text(4, 2000, ' tabla 15x15 4 peones') 
  text(5, 3000, 'tabla 25x25 2 peones') 
  text(6, 3500, ' tabla 25x 25 4 peones') 
  grid()
  legend(loc='upper right')
  
  show()
```

```
Pdv = [0, 89.33368945121765, 44.1855263710022, 98.3169014453888, 98.19989824295044, 229.0978240966797, 220.007994890213]
PBFS = [0, 565.314531326294, 584.627628326416, 1354.5148372650146, 1435.95552444458, 3494.032144546509, 3702.0716667175293]

Grafico(Pdv, PBFS)
```
